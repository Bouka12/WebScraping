# Setup

```python
import requests 
from bs4 import BeautifulSoup
```

```python
base_site = "https://en.wikipedia.org/wiki/List_of_national_capitals_by_population"
```

```python
r = requests.get(base_site)
```

```python
html = r.content
```

# Extracting tables with Beautiful Soup

```python
soup = BeautifulSoup(html, 'lxml')
```

```python
soup
```

```python
table_ = soup.find_all('table')
table_[3]
```

```python
table = soup.find_all('table')[1]
```

```python
table
```

```python
table.find_all("tr")[0].contents
```

```python
capitals = [row.contents[3].text for row in table.find_all("tr")]
capitals
```

# Using Pandas to extract tables

* `pandas.read_html(HTML/URL)`: it returns a list of all tables on the page as Data Frames

```python
import pandas as pd
```

```python
tables = pd.read_html(base_site)
```

```python
type(tables[0])
```
