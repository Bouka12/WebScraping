# Set-up

```python
import requests
from bs4 import BeautifulSoup
```

```python
base_site = "https://editorial.rottentomatoes.com/guide/best-netflix-shows-and-movies-to-binge-watch-now/"
```

```python
response = requests.get(base_site)
response
```

```python
html = response.content
```

# Choosing a parser

```python
soup = BeautifulSoup(html, 'html.parser') 
# 'lxml'
```

```python
with open('Rotten_tomatoes_bestNeflix.html','wb') as file:
    file.write(soup.prettify('utf-8'))
    
```

# Obtaining the element containing all the data

```python
divs = soup.find_all('div', {'class': 'row countdown-item'})
divs
```

# Extracting the title and year of each movie

```python
#divs_ = [div for div in divs if div is not None]
headings =[div.find('h2') for div in divs]
headings
```

```python
headings[0]
```

## Title

```python
movie_names = [heading.find('a').string for heading in headings]
movie_names
```

## Years

```python
years = [heading.find('span', class_ ='start-year').string for heading in headings]
years
```

### Removing the brackets

```python
years[0][1:-1]
```

* `.strip(symbols)` method:
It removes all leading and trainling symbols from a string. The symbols to be removed are supplied as a parameter

```python
# .strip(symbols)
years[0].strip('()')
```

```python
years = [year.strip('()') for year  in years]
years
```

```python
years = [int(year) for year in years]
years
```

## Score

```python
scores = [heading.find('span', class_='tMeterScore').string for heading in headings]
scores
```

```python
scores = [score.strip('%') for score in scores]
scores
```

```python
scores = [int(score) for score in scores]
scores
```

# Extracting the rest of the information

## Synopsis

```python
consensus = [div.find('div', class_="info synopsis") for div in divs]
consensus
```

```python
consensus[0]
```

```python
[cons.text for cons in consensus]
```

```python
common_text = '\nSynopsis: '
```

```python
len(common_text)
```

```python
start = len(common_text)
```

```python
end_word = '\n               [More]\n'
len(end_word)
```

```python
end=len(end_word)
```

```python
consensus_text = [cons.text[start:-end] for cons in consensus]
consensus_text
```

## Directors

```python
divs[0]
```

```python
directors = [div.find('div', {'class': 'info director'}) for div in divs]
directors
```

```python
directors = [None if dire.find("a")is None else dire.find("a").string  for dire in directors]
directors
```

```python
directors[0].find_all('a')
```

```python
# for loops
all_names = []
for director in directors:
    l_directs = director.find_all('a')
    l_names=[l_direct.string for l_direct in l_directs]
    res = ", ".join(l_names)
    all_names.append(res)
all_names
```

```python
# nested loops
dir_names = [", ".join([l_direct.string for l_direct in director.find_all('a')]) for director in directors]
dir_names
```

## Cast info

```python
divs[0]
```

```python
# class = "info cast"
cast_info = [div.find('div', {'class':'info cast'}) for div in divs]
cast_info
```

```python
cast_info[0]
```

```python
cast_info[0].find_all('a')
```

```python
cast_links = cast_info[0].find_all('a')
cast_links
```

```python
cast_names = [link.string for link in cast_links]
cast_names
```

```python
cast = ", ".join(cast_names)
cast
```

### Using a 'for loop'

```python
cast = []
for c in cast_info:
    cast_links = c.find_all('a')
    cast_names = [link.string for link in cast_links]
    result = ", ".join(cast_names)
    cast.append(result)
cast
```

### Nested list comprehension

```python
cast = [", ".join([link.string for link in c.find_all('a')]) for c in cast_info]
cast
```

## Representing the data in structured form

```python
import pandas as pd
```

### Creating a Data Frame and populating it

```python
movies_info = pd.DataFrame()

movies_info["Movie Title"] = movie_names
movies_info["Year"] = years
movies_info["Score"] = scores
#movie_info["Adjusted Score"] = 
movies_info["Directors"] = dir_names
#movie_info["Synopsis"] = synopsis_text
movies_info["Cast"] = cast
movies_info["Consensus"] = consensus_text

movies_info.head()
```

```python
pd.set_option("display.max_colwidth", 1)
movies_info
```

### Exporting the data to CSV and Excel files

```python
movies_info.to_csv("movies_info.csv", index=False, header =True)
```
