# CSS Selectors

* CSS selectors is a notation for filtering certain HTML elements, a.k.a. tags

## Select elements based on tag name

```python
# tag name
r.html.find("span")
```

## Select elements based upon ID

```python
r.html.find("#Name")
```

```python
r.html.find("#name") # Case sensitive
# ID  is unique one or no result
```

```python
r.html.find("#Name", first=True)
```

## Selecting by class

```python
r.html.find(".reference")
```

```python
r.html.find(".mw-file-description")
```

```python
r.html.find(".metadata")
```

```python
# in case of selecting multiple classes we pass ".class1.class2"
```

## Selecting based on other attribute

* **Reference** on CSS selectors:
`www.w3schools.com/cssref/css_selectors.asp`

```python
# [attribute] selects all tags that have defines the attribute
r.html.find("[target]")
```

```python
# [attribute=value] -- selects all tags with that particular value of the attribute
r.html.find("[role=note]")
```

```python
# [attribute*=value] -- attribute contains the SUBSTRING 'value'
r.html.find("[href*=wikipedia]")
```

```python
r.html.find("[href^=http]") # starting with http
```

## Combining different filters together

```python

r.html.find("a[href*=wikipedia]")
```

```python
r.html.find("a.internal")
```

```python
r.html.find("div.thumb.tright")
```

```python
r.html.find("div[role=note][class='hatnote navigation-not-searchable']")
```

## Incorporating tag hierarchy

```python
r.html.find("h1 span")
```

```python
r.html.find("div > p") # parent is div # child is p
```
